```python
# bargraph
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
ola_days=[0,1,2,3,4,5]
ola_km=[10,22,40,50,60,90]
uber_days=[4,5,6,7,8]
uber_km=[50,60,80,100,110]
plt.bar(ola_days,ola_km,label="OLA cabs")
plt.bar(uber_days,uber_km,label="UBER cabs")
plt.title("UBER Vs OLA")
plt.legend()
plt.xlabel('no. of days')
plt.ylabel('no of km/day')
plt.show()
```

### Box Plot: Used for displaying of data base on min ,IQR,max values
- min
- IQR 
  * lower Quartile(25%)
  * Median(50%)
  * Upper Quartile(75%)
- max

```python
marks = np.random.randint(20,100,101)
marks
```

```python
plt.boxplot(marks,patch_artist=True)
plt.show()
```

```python
v1=np.random.randint(50,100,10)
v2 = np.random.randint(40,90,10)
v3=np.random.randint(30,100,10)
```

```python
v1 ,v2,v3
```

```python
plt.boxplot([v1,v2,v3],labels=["value1","value2","value3"])
plt.show()
```

### Area Plot

```python
days = [1,2,3,4]
working = [8,11,9,10]
sleeping = [2,10,11,12]
plt.stackplot(days,working,sleeping,colors=['g','r'],labels=["work","rest"])
plt.legend()
plt.show()
```
